```python
# Exemplo prÃ¡tico: CÃ¡lculo de juros compostos
# CenÃ¡rio: Investimento em CDB

capital_inicial = 10000  # R$ 10.000
taxa_anual = 0.12        # 12% ao ano (CDI mÃ©dio)
anos = 5

# FÃ³rmula: M = C * (1 + i)^t
montante_final = capital_inicial * (1 + taxa_anual) ** anos
rendimento = montante_final - capital_inicial
rentabilidade_percentual = (rendimento / capital_inicial) * 100

print("ðŸ’° SIMULAÃ‡ÃƒO DE INVESTIMENTO EM CDB")
print("-" * 50)
print(f"Capital inicial:     R$ {capital_inicial:>12,.2f}")
print(f"Taxa anual:          {taxa_anual:>15.1%}")
print(f"PerÃ­odo:             {anos:>15} anos")
print("-" * 50)
print(f"Montante final:      R$ {montante_final:>12,.2f}")
print(f"Rendimento:          R$ {rendimento:>12,.2f}")
print(f"Rentabilidade:       {rentabilidade_percentual:>15.1f}%")
print("-" * 50)
```

```python
# Visualizar crescimento do capital ao longo do tempo
import os

anos_array = np.arange(0, anos + 1)
montantes = capital_inicial * (1 + taxa_anual) ** anos_array

# Configurar estilo
plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(12, 6))

# GrÃ¡fico de linha
plt.plot(anos_array, montantes, 
         marker='o', 
         linewidth=3, 
         markersize=10,
         color='#2E86AB',
         label='Montante')

# Linha do capital inicial (referÃªncia)
plt.axhline(y=capital_inicial, 
            color='red', 
            linestyle='--', 
            linewidth=2,
            alpha=0.7,
            label='Capital Inicial')

# CustomizaÃ§Ã£o
plt.title('ðŸ“ˆ Crescimento do Capital com Juros Compostos (12% a.a.)', 
          fontsize=16, 
          fontweight='bold',
          pad=20)
plt.xlabel('Anos', fontsize=12, fontweight='bold')
plt.ylabel('Montante (R$)', fontsize=12, fontweight='bold')
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)

# AnotaÃ§Ãµes nos pontos
for i, (ano, valor) in enumerate(zip(anos_array, montantes)):
    plt.annotate(f'R$ {valor:,.0f}', 
                 xy=(ano, valor), 
                 xytext=(0, 10),
                 textcoords='offset points',
                 ha='center',
                 fontsize=9,
                 bbox=dict(boxstyle='round,pad=0.3', 
                          facecolor='yellow', 
                          alpha=0.3))

plt.tight_layout()
plt.show()

print(f"\nâœ… GrÃ¡fico gerado com sucesso!")
print(f"Em {anos} anos, seu capital cresceu {rentabilidade_percentual:.1f}%")
```

```python
# Dados recebidos de um sistema externo (tipos bagunÃ§ados!)
dados_misturados = {
    'faturamento': "125000",      # Veio como string, deveria ser int
    'margem': "0.15",             # Veio como string, deveria ser float
    'ativo': "True",              # Veio como string, deveria ser bool
    'funcionarios': 42,           # âœ… Ãšnico correto (int)
    'fundacao': "2020"            # Veio como string, deveria ser int
}

# VERIFICAR: Tipos atuais (errados)
print("ðŸ”´ TIPOS ATUAIS (INCORRETOS):")
print("=" * 60)
for chave, valor in dados_misturados.items():
    tipo = type(valor).__name__
    print(f"{chave:15s} â†’ {tipo:10s} | Valor: {valor}")

# CORRIGIR: Converter para os tipos corretos
dados_corretos = {
    'faturamento': int(dados_misturados['faturamento']),           # "125000" â†’ 125000
    'margem': float(dados_misturados['margem']),                   # "0.15" â†’ 0.15
    'ativo': dados_misturados['ativo'].lower() == "true",          # "True" â†’ True (bool)
    'funcionarios': dados_misturados['funcionarios'],              # 42 (jÃ¡ estÃ¡ correto)
    'fundacao': int(dados_misturados['fundacao'])                  # "2020" â†’ 2020
}

print("\n" + "=" * 60)
print("ðŸŸ¢ TIPOS CORRIGIDOS:")
print("=" * 60)
for chave, valor in dados_corretos.items():
    tipo = type(valor).__name__
    print(f"{chave:15s} â†’ {tipo:10s} | Valor: {valor}")

# USAR OS DADOS CORRIGIDOS: Calcular lucro
lucro = dados_corretos['faturamento'] * dados_corretos['margem']
idade_empresa = 2025 - dados_corretos['fundacao']

print("\n" + "=" * 60)
print("ðŸ’° ANÃLISE DA EMPRESA")
print("=" * 60)
print(f"Faturamento:       R$ {dados_corretos['faturamento']:>12,.2f}")
print(f"Margem de lucro:   {dados_corretos['margem']:>15.1%}")
print(f"Lucro estimado:    R$ {lucro:>12,.2f}")
print(f"FuncionÃ¡rios:      {dados_corretos['funcionarios']:>15} pessoas")
print(f"Fundada em:        {dados_corretos['fundacao']:>19}")
print(f"Idade da empresa:  {idade_empresa:>15} anos")
print(f"Status:            {'ðŸŸ¢ ATIVA' if dados_corretos['ativo'] else 'ðŸ”´ INATIVA'}")
print("=" * 60)
```
