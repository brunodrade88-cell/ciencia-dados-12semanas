```python
# Dados de uma empresa fictícia (TechCorp)
receita_bruta = 850000
impostos_sobre_vendas = 127500
devolucoes = 42500
cmv = 425000
despesas_operacionais = 170000
despesas_financeiras = 25500

# VOCÊ FAZ AGORA: Calcule cada linha abaixo
# Dica: use as fórmulas nos comentários!

# 1. Receita Líquida = Receita Bruta - Impostos - Devoluções
receita_liquida = receita_bruta - impostos_sobre_vendas - devolucoes

# 2. Lucro Bruto = Receita Líquida - CMV
lucro_bruto = receita_liquida - cmv

# 3. EBITDA = Lucro Bruto - Despesas Operacionais
ebitda = lucro_bruto - despesas_operacionais

# 4. Lucro Líquido = EBITDA - Despesas Financeiras
lucro_liquido = ebitda - despesas_financeiras

# 5. Margem Bruta (%) = (Lucro Bruto / Receita Líquida) * 100
margem_bruta = (lucro_bruto / receita_liquida) * 100

# 6. Margem Líquida (%) = (Lucro Líquido / Receita Líquida) * 100
margem_liquida = (lucro_liquido / receita_liquida) * 100

# Imprimir resultados
print("="*60)
print("DEMONSTRAÇÃO DO RESULTADO DO EXERCÍCIO (DRE)")
print("="*60)
print(f"Receita Bruta:          R$ {receita_bruta:>15,.2f}")
print(f"(-) Impostos:           R$ {impostos_sobre_vendas:>15,.2f}")
print(f"(-) Devoluções:         R$ {devolucoes:>15,.2f}")
print(f"Receita Líquida:        R$ {receita_liquida:>15,.2f}")
print(f"\n(-) CMV:                R$ {cmv:>15,.2f}")
print(f"Lucro Bruto:            R$ {lucro_bruto:>15,.2f}")
print(f"    Margem Bruta:       {margem_bruta:>18.2f}%")
print(f"\n(-) Desp. Operacionais: R$ {despesas_operacionais:>15,.2f}")
print(f"EBITDA:                 R$ {ebitda:>15,.2f}")
print(f"\n(-) Desp. Financeiras:  R$ {despesas_financeiras:>15,.2f}")
print(f"Lucro Líquido:          R$ {lucro_liquido:>15,.2f}")
print(f"    Margem Líquida:     {margem_liquida:>18.2f}%")
print("="*60)
```

```python
# Notas fiscais bagunçadas (formato brasileiro real)
nf1 = "NF-e 000045821 | VALOR: R$ 15.247,89 | EMITENTE: ACME Corp. S.A."
nf2 = "NF-e 000045822 | VALOR: R$ 8.129,50 | EMITENTE: Beta Solutions Ltda."
nf3 = "NF-e 000045823 | VALOR: R$ 23.841,00 | EMITENTE: Gamma Tech EIRELI"

# Função que limpa UMA nota fiscal
def processar_nota(nf_string):
    """Extrai: número, valor, emitente de uma string de NF"""
    
    # Dividir em 3 partes pelo " | "
    partes = nf_string.split(" | ")
    # Resultado: ['NF-e 000045821', 'VALOR: R$ 15.247,89', 'EMITENTE: ACME Corp. S.A.']
    
    # EXTRAIR NÚMERO DA NOTA
    numero_str = partes[0].split()[1]  # Pega "000045821" (segunda palavra)
    numero = int(numero_str)            # Converte para número: 45821
    
    # EXTRAIR VALOR (converter formato brasileiro → float)
    valor_str = partes[1].replace("VALOR: R$ ", "")  # Remove prefixo → "15.247,89"
    valor_str = valor_str.replace(".", "")             # Remove ponto de milhar → "15247,89"
    valor_str = valor_str.replace(",", ".")            # Troca vírgula por ponto → "15247.89"
    valor = float(valor_str)                           # Converte para float: 15247.89
    
    # EXTRAIR EMITENTE (limpar sufixos empresariais)
    emitente = partes[2].replace("EMITENTE: ", "")
    emitente = emitente.replace(" S.A.", "")
    emitente = emitente.replace(" Ltda.", "")
    emitente = emitente.replace(" EIRELI", "")
    
    return numero, valor, emitente

# Processar as 3 notas fiscais
n1, v1, e1 = processar_nota(nf1)
n2, v2, e2 = processar_nota(nf2)
n3, v3, e3 = processar_nota(nf3)

# Exibir resultados formatados
print("=" * 60)
print("NOTAS FISCAIS PROCESSADAS")
print("=" * 60)
print(f"NF {n1:06d} | R$ {v1:>12,.2f} | {e1}")
print(f"NF {n2:06d} | R$ {v2:>12,.2f} | {e2}")
print(f"NF {n3:06d} | R$ {v3:>12,.2f} | {e3}")
print("=" * 60)
print(f"TOTAL:      | R$ {v1+v2+v3:>12,.2f} |")
print("=" * 60)
```
