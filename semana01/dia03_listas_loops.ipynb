# üìä Dia 3: Listas, Tuplas e Loops

## Objetivos
- Manipular listas de dados financeiros
- Iterar com `for` e `while`
- Calcular m√©tricas de s√©ries temporais
- Usar list comprehension

## Data
12/11/2025

## Contexto
Trabalhar com dados de m√∫ltiplos per√≠odos (receitas mensais, cota√ß√µes)

---

```python
# Dados: Receitas mensais de uma empresa (em milhares de R$)
meses = ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun", 
         "Jul", "Ago", "Set", "Out", "Nov", "Dez"]

receitas = [450, 520, 480, 550, 600, 580, 
            620, 590, 610, 640, 630, 650]

print("=" * 60)
print("AN√ÅLISE DE RECEITAS - 2025")
print("=" * 60)

# 1. ESTAT√çSTICAS B√ÅSICAS
total_anual = sum(receitas)
media_mensal = total_anual / len(receitas)
maior_receita = max(receitas)
menor_receita = min(receitas)

print(f"\nüìä ESTAT√çSTICAS GERAIS")
print(f"Total anual:        R$ {total_anual:>10,.0f}k")
print(f"M√©dia mensal:       R$ {media_mensal:>10,.2f}k")
print(f"Maior receita:      R$ {maior_receita:>10,.0f}k")
print(f"Menor receita:      R$ {menor_receita:>10,.0f}k")

# 2. ENCONTRAR MESES DE PICO E VALE
indice_maior = receitas.index(maior_receita)
indice_menor = receitas.index(menor_receita)

print(f"\nüîù Melhor m√™s:      {meses[indice_maior]} (R$ {maior_receita}k)")
print(f"üìâ Pior m√™s:        {meses[indice_menor]} (R$ {menor_receita}k)")

# 3. CRESCIMENTO M√äS A M√äS
print(f"\nüìà CRESCIMENTO M√äS A M√äS:")
for i in range(1, len(receitas)):
    mes_atual = meses[i]
    receita_atual = receitas[i]
    receita_anterior = receitas[i-1]
    
    variacao = receita_atual - receita_anterior
    percentual = (variacao / receita_anterior) * 100
    
    simbolo = "üü¢" if variacao >= 0 else "üî¥"
    print(f"{simbolo} {mes_atual}: {variacao:>+6.0f}k ({percentual:>+6.2f}%)")

print("=" * 60)
```

## üí∞ EXERC√çCIO 1: An√°lise de Fluxo de Caixa Semanal

**Objetivo:** Praticar listas, loops e c√°lculos iterativos

**Dados:** Fluxo de caixa di√°rio de uma semana (seg a dom)

**Tarefas:**
1. Calcular saldo di√°rio (entradas - sa√≠das)
2. Calcular saldo acumulado dia a dia
3. Identificar dias com saldo negativo
4. Calcular m√©dias
5. Encontrar melhor e pior dia

---

```python
# Fluxo de caixa di√°rio de uma semana (em R$)
dias = ["Seg", "Ter", "Qua", "Qui", "Sex", "S√°b", "Dom"]
entradas = [12500, 15800, 14200, 16500, 18900, 22000, 8500]
saidas = [10200, 13500, 12800, 14200, 15800, 18500, 6200]

# 1. SALDO DI√ÅRIO (usando list comprehension)
saldos = [entradas[i] - saidas[i] for i in range(len(entradas))]

# 2. SALDO ACUMULADO (usando loop)
saldo_acumulado = []
total = 0
for saldo in saldos:
    total += saldo
    saldo_acumulado.append(total)

# 3. DIAS COM SALDO NEGATIVO (list comprehension com filtro)
dias_negativos = [dias[i] for i, s in enumerate(saldos) if s < 0]

# 4. M√âDIAS
media_entradas = sum(entradas) / len(entradas)
media_saidas = sum(saidas) / len(saidas)

# 5. MELHOR E PIOR DIA
melhor_indice = saldos.index(max(saldos))
pior_indice = saldos.index(min(saldos))

# ===== EXIBIR RESULTADOS =====
print("=" * 70)
print("FLUXO DE CAIXA SEMANAL")
print("=" * 70)

# Tabela formatada
print(f"\n{'Dia':^6} | {'Entradas':>10} | {'Sa√≠das':>10} | {'Saldo':>10} | {'Acumulado':>12}")
print("-" * 70)

for i in range(len(dias)):
    print(f"{dias[i]:^6} | R$ {entradas[i]:>7,.2f} | R$ {saidas[i]:>7,.2f} | "
          f"R$ {saldos[i]:>7,.2f} | R$ {saldo_acumulado[i]:>9,.2f}")

print("-" * 70)
print(f"{'TOTAL':^6} | R$ {sum(entradas):>7,.2f} | R$ {sum(saidas):>7,.2f} | "
      f"R$ {sum(saldos):>7,.2f} | R$ {saldo_acumulado[-1]:>9,.2f}")

# An√°lises
print(f"\nüìä AN√ÅLISES:")
print(f"M√©dia de entradas:  R$ {media_entradas:>10,.2f}")
print(f"M√©dia de sa√≠das:    R$ {media_saidas:>10,.2f}")
print(f"\nüîù Melhor dia:      {dias[melhor_indice]} (saldo R$ {saldos[melhor_indice]:,.2f})")
print(f"üìâ Pior dia:        {dias[pior_indice]} (saldo R$ {saldos[pior_indice]:,.2f})")

if dias_negativos:
    print(f"\n‚ö†Ô∏è  Dias negativos:  {', '.join(dias_negativos)}")
else:
    print(f"\n‚úÖ Todos os dias positivos!")

print("=" * 70)
```

## üìà EXERC√çCIO 2: Simula√ß√£o de Investimento com Aportes

**Objetivo:** Usar loop `while` para simular crescimento at√© atingir meta

**Cen√°rio:**
- Capital inicial: R$ 10.000
- Meta: R$ 20.000
- Taxa mensal: 1.5%
- Aporte mensal: R$ 500

**Perguntas:**
- Quantos meses para atingir a meta?
- Quanto rendeu?
- Qual foi o total investido?

---

```python
# Configura√ß√£o da simula√ß√£o
capital_inicial = 10000
meta = 20000
taxa_mensal = 0.015  # 1.5% ao m√™s
aporte_mensal = 500

# Inicializar vari√°veis
capital = capital_inicial
meses = 0
evolucao = [capital]  # Lista para armazenar evolu√ß√£o

print("=" * 60)
print("SIMULA√á√ÉO DE INVESTIMENTO COM APORTES")
print("=" * 60)
print(f"Capital inicial:  R$ {capital_inicial:>10,.2f}")
print(f"Meta:             R$ {meta:>10,.2f}")
print(f"Taxa mensal:      {taxa_mensal:>13.2%}")
print(f"Aporte mensal:    R$ {aporte_mensal:>10,.2f}")
print("=" * 60)

# Loop while - simula m√™s a m√™s at√© atingir meta
while capital < meta:
    meses += 1
    capital = capital * (1 + taxa_mensal)  # Aplica rendimento
    capital += aporte_mensal                # Adiciona aporte
    evolucao.append(capital)
    
    # Mostrar primeiros 5 meses e depois a cada 6 meses
    if meses <= 5 or meses % 6 == 0:
        print(f"M√™s {meses:>3}: R$ {capital:>10,.2f}")

# C√°lculos finais
total_investido = capital_inicial + (aporte_mensal * meses)
rendimento_total = capital - total_investido
rentabilidade = (rendimento_total / total_investido) * 100

print("=" * 60)
print(f"\n‚úÖ META ATINGIDA EM {meses} MESES!")
print(f"\nValor final:      R$ {capital:>10,.2f}")
print(f"Total investido:  R$ {total_investido:>10,.2f}")
print(f"Rendimento:       R$ {rendimento_total:>10,.2f}")
print(f"Rentabilidade:    {rentabilidade:>13.2f}%")
print("=" * 60)

# Mostrar evolu√ß√£o nos √∫ltimos 3 meses
print(f"\nüìä EVOLU√á√ÉO DOS √öLTIMOS 3 MESES:")
for i in range(-3, 0):
    mes_num = meses + i + 1
    print(f"M√™s {mes_num}: R$ {evolucao[i]:,.2f}")
```

```python
# Simula√ß√£o: Quanto tempo para duplicar investimento?
capital = 1000
taxa_anual = 0.10  # 10% ao ano
anos = 0

print(f"Capital inicial: R$ {capital:,.2f}")
print(f"Taxa: {taxa_anual:.0%} ao ano")
print(f"Meta: Duplicar o capital\n")

while capital < 2000:
    anos += 1
    capital = capital * (1 + taxa_anual)
    print(f"Ano {anos}: R$ {capital:,.2f}")

print(f"\nüéØ Levou {anos} anos para duplicar!")
print(f"Regra dos 72: ~{72/10:.0f} anos (aproxima√ß√£o)")
```

## üèÜ EXERC√çCIO 3: Ranking de Empresas com List Comprehension

**Objetivo:** Dominar list comprehension e opera√ß√µes com listas

**Dados:** 5 empresas com receitas e despesas

**Tarefas:**
1. Calcular lucro de cada empresa
2. Calcular margem l√≠quida (%)
3. Filtrar empresas com margem > 15%
4. Criar ranking ordenado
5. Identificar campe√£s e deficit√°rias

---

```python
# Dados de 5 empresas do setor de tecnologia
empresas = ["TechCorp", "FinSolutions", "DataPro", "CloudNet", "AlphaTech"]
receitas = [1500000, 850000, 2200000, 950000, 1800000]
despesas = [1200000, 720000, 1850000, 880000, 1500000]

print("=" * 75)
print("AN√ÅLISE COMPARATIVA DE EMPRESAS - SETOR TECH")
print("=" * 75)

# 1. CALCULAR LUCROS (list comprehension)
lucros = [receitas[i] - despesas[i] for i in range(len(receitas))]

# 2. CALCULAR MARGENS L√çQUIDAS (%)
margens = [(lucros[i] / receitas[i]) * 100 for i in range(len(receitas))]

# 3. FILTRAR EMPRESAS COM MARGEM > 15%
# (list comprehension com condicional)
boas_margens = [empresas[i] for i, m in enumerate(margens) if m > 15]

# 4. CRIAR LISTA DE TUPLAS (nome, lucro, margem)
dados_completos = [
    (empresas[i], lucros[i], margens[i]) 
    for i in range(len(empresas))
]

# 5. ORDENAR POR MARGEM (decrescente)
# lambda = fun√ß√£o an√¥nima que pega elemento x e retorna x[2] (margem)
ranking = sorted(dados_completos, key=lambda x: x[2], reverse=True)

# ===== EXIBIR RESULTADOS =====

# Tabela principal
print(f"\n{'#':^4} | {'Empresa':^15} | {'Receita':>15} | {'Despesas':>15} | {'Lucro':>15} | {'Margem':>10}")
print("-" * 75)

for i, (nome, lucro, margem) in enumerate(ranking, 1):
    # Buscar receita e despesa original
    idx_original = empresas.index(nome)
    receita = receitas[idx_original]
    despesa = despesas[idx_original]
    
    # Emoji para o ranking
    if i == 1:
        emoji = "ü•á"
    elif i == 2:
        emoji = "ü•à"
    elif i == 3:
        emoji = "ü•â"
    else:
        emoji = "  "
    
    # Status da margem
    if margem > 15:
        status = "‚úÖ"
    elif margem > 10:
        status = "‚ö†Ô∏è "
    else:
        status = "üî¥"
    
    print(f"{emoji}{i:>2} | {nome:^15} | R$ {receita:>12,.2f} | R$ {despesa:>12,.2f} | R$ {lucro:>12,.2f} | {status} {margem:>6.2f}%")

print("-" * 75)

# Estat√≠sticas gerais
print(f"\nüìä ESTAT√çSTICAS DO SETOR:")
print(f"Receita total:      R$ {sum(receitas):>15,.2f}")
print(f"Despesa total:      R$ {sum(despesas):>15,.2f}")
print(f"Lucro total:        R$ {sum(lucros):>15,.2f}")
print(f"Margem m√©dia:       {sum(margens)/len(margens):>18.2f}%")

# Destaques
print(f"\nüèÜ DESTAQUES:")
print(f"Maior lucro:        {ranking[0][0]} (R$ {ranking[0][1]:,.2f})")
print(f"Melhor margem:      {ranking[0][0]} ({ranking[0][2]:.2f}%)")
print(f"Pior margem:        {ranking[-1][0]} ({ranking[-1][2]:.2f}%)")

# Empresas com boa margem
if boas_margens:
    print(f"\n‚úÖ Empresas com margem > 15% ({len(boas_margens)}):")
    print(f"   {', '.join(boas_margens)}")
else:
    print(f"\n‚ö†Ô∏è  Nenhuma empresa com margem > 15%")

# Empresas deficit√°rias
deficitarias = [empresas[i] for i, l in enumerate(lucros) if l < 0]
if deficitarias:
    print(f"\nüî¥ Empresas com preju√≠zo:")
    print(f"   {', '.join(deficitarias)}")

print("=" * 75)
```

```python
# Lista de empresas com dados
empresas_dados = [
    ("TechCorp", 1500000, 1200000),
    ("DataPro", 2200000, 1850000),
    ("CloudNet", 950000, 880000)
]

# Lambda para calcular lucro
calcular_lucro = lambda emp: emp[1] - emp[2]

# Lambda para calcular margem
calcular_margem = lambda emp: ((emp[1] - emp[2]) / emp[1]) * 100

print("AN√ÅLISE COM LAMBDAS:\n")

for empresa in empresas_dados:
    nome = empresa[0]
    lucro = calcular_lucro(empresa)
    margem = calcular_margem(empresa)
    
    print(f"{nome:12s} | Lucro: R$ {lucro:>10,.2f} | Margem: {margem:>6.2f}%")

# Ordenar por margem (usando lambda direto no sorted)
print("\n" + "="*50)
print("RANKING POR MARGEM:")
print("="*50)

ranking_margem = sorted(empresas_dados, 
                        key=lambda x: calcular_margem(x), 
                        reverse=True)

for i, empresa in enumerate(ranking_margem, 1):
    nome = empresa[0]
    margem = calcular_margem(empresa)
    emoji = "ü•á" if i == 1 else "ü•à" if i == 2 else "ü•â"
    print(f"{emoji} {i}¬∫ - {nome}: {margem:.2f}%")
```
